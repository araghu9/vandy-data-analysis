```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
dfCrit=pd.read_csv("data/cleaned_datasets-165618-377201-metacritic_games.csv", index_col=1)
dfSale=pd.read_csv("data/cleaned_vgsales.csv", index_col=1)
```

```python
genre=dfCrit["genre"]
count = dfCrit['genre'].value_counts().tolist()
del count[30:52]
fig = plt.figure(figsize=(20, 5))
plt.xticks(rotation=90)
plt.xlabel("Genres")
plt.ylabel("Metascore")
plt.title("Metascores of various games in different genres")
plt.scatter(genre,dfCrit["metascore"])
```

```python
genre=dfSale["genre"]
count = dfSale['genre'].value_counts().tolist()
del count[30:52]
fig = plt.figure(figsize=(20, 5))
plt.xticks(rotation=90)
plt.xlabel("Genres")
plt.ylabel("Global Sales(USD in millions)")
plt.title("Total revenue of various games in different genres")
plt.scatter(genre,dfSale["global_sales"])
```

```python
genre=dfCrit["genre"]
genre = dfCrit['genre'].value_counts().keys().tolist()
count = dfCrit['genre'].value_counts().tolist()
del genre[35:52]
del count[35:52]
fig = plt.figure(figsize=(20, 13))
plt.xticks(rotation=90)
plt.xlabel("Genres")
plt.ylabel("Number of Games")
plt.title("Number of games in different genres")
plt.bar(genre,count)
```
